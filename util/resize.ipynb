```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import albumentations as A
import os
from pathlib import Path
from PIL import Image
```

```python
csv_path = '/opt/ml/input/data/eval/info.csv'
# train_dir = True
# target_dir = 'train' if train_dir else 'eval'
# resize_path = '/opt/ml/input/data/{}/images_resize'.format(target_dir)

# if not os.path.isdir(resize_path):
#     os.mkdir(resize_path)

df = pd.read_csv(csv_path)
print(len(df))
print(df['ImageID'])
```

```python
ori_path = '/opt/ml/input/data/eval/images'
resize_path = '/opt/ml/input/data/eval/images_resize'
```

```python
count = 0

for f_name in df['ImageID']:
    o_path = Path(ori_path) / Path(f_name)
    
    img = Image.open(o_path)
    
    transform = A.Compose([
                        A.Resize(224, 224)
                        ])
    img = Image.fromarray(transform(image=np.array(img))['image'])
    new_path = Path(resize_path) / f_name
    img.save(new_path)
    count += 1
    if(count % 100 == 0):
        print(count)
```
